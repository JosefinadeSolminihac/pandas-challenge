```python
#Dependencies
import pandas as pd
```

```python
#Save path to the data sets in a variable
school_data="../Resources/schools_complete.csv"
students_data="../Resources/students_complete.csv"
```

```python
#Use pandas to read data
school_df=pd.read_csv(school_data)
students_df=pd.read_csv(students_data)
```

```python
# Combine the data into a single dataset
school_data_complete = pd.merge(students_df, school_df, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
#Total number of unique schools
school_count=len(school_data_complete["school_name"].unique())
school_count
```

```python
#Total students
total_students=school_data_complete["Student ID"].count()
total_students
```

```python
#Total budget: convert /Using the data not merged -> total_budget=school_df["budget"].sum() total_budget
total_budget=(school_data_complete["budget"].unique()).sum()
total_budget
```

```python
#Average math score
average_math=school_data_complete["math_score"].mean()
average_math
```

```python
#Average reading score
average_reading=school_data_complete["reading_score"].mean()
average_reading
```

```python
#% passing math (the percentage of students who passed math)(math scores greather than or equal to 70)
students_passing_math=school_data_complete[(school_data_complete["math_score"]>=70)].count()["student_name"]
percentage_passing_math=students_passing_math/total_students*100
percentage_passing_math
```

```python
#% passing reading (the percentage of students who passed reading)
students_passing_reading=school_data_complete[(school_data_complete["reading_score"]>=70)].count()["student_name"]
percentage_passing_reading=students_passing_reading/total_students*100
percentage_passing_reading
```

```python
#% overall passing (the percentage of students who passed math AND reading)
students_passing_math_reading=school_data_complete.loc[(school_data_complete["math_score"]>=70)&(school_data_complete["reading_score"]>=70)].count()["student_name"]
percentage_passing_math_reading=students_passing_math_reading/total_students*100
percentage_passing_math_reading
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
summary_district=pd.DataFrame([{
    "Total Schools":school_count,
    "Total Students":total_students,
    "Total Budget": total_budget,
    "Average Math Score":average_math,
    "Average Reading Score":average_reading,
    "% Passing Math":percentage_passing_math, 
    "% Passing Math":percentage_passing_reading,
    "% Overall Passing":percentage_passing_math_reading
}])

#Give format 
summary_district["Total Students"]=summary_district["Total Students"].map("{:,}".format)
summary_district["Total Budget"]=summary_district["Total Budget"].map("${:,.2f}".format)

summary_district
```

```python
#SCHOOL SUMMARY: Perform the necessary calculations and then create a DataFrame that summarizes key metrics about each school.
#Set the DataFrame index using existing columns. For these case: the school name and its type.
school_types = school_df.set_index(["school_name"])["type"]
df=pd.DataFrame(school_types)
df
```

```python
#Calculate the total student per school
total_students_school=school_data_complete.groupby(["school_name"]).count()["Student ID"]
total_students_school
```

```python
# Calculate the total school budget
per_school_budget = school_data_complete.groupby(["school_name"]).mean()["budget"]
per_school_budget
```

```python
# Calculate Per student budget
per_school_capita = per_school_budget / total_students_school
per_school_capita
```

```python
# Calculate the average math test scores
average_math_score_school=school_data_complete.groupby(["school_name"]).mean()["math_score"]
average_math_score_school
```

```python
# Calculate the average reading test scores
average_reading_score_school=school_data_complete.groupby(["school_name"]).mean()["reading_score"]
average_reading_score_school
```

```python
#Calculate % passing math (the percentage of students who passed math)
passing_math=school_data_complete.loc[school_data_complete["math_score"]>=70]
percentage_passing_math_school=passing_math.groupby(["school_name"]).count()["student_name"]/total_students_school*100
percentage_passing_math_school
```

```python
#Calculate % passing reading (the percentage of students who passed reading)
passing_reading=school_data_complete.loc[school_data_complete["reading_score"]>=70]
percentage_passing_reading_school=passing_reading.groupby(["school_name"]).count()["student_name"]/total_students_school*100
percentage_passing_reading_school
```

```python
#% overall passing (the percentage of students who passed math AND reading)
overall_passing=school_data_complete.loc[(school_data_complete["math_score"]>=70)&(school_data_complete["reading_score"]>=70),:]
overall_passing_rate=overall_passing.groupby(["school_name"]).count()["student_name"] / total_students_school * 100
overall_passing_rate
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.
summary_schools=pd.DataFrame({
    "School Type":school_types,
    "Total Students":total_students_school,
    "Total School Budget":per_school_budget,
    "Per Student Budget":per_school_capita,
    "Average Math Score":average_math_score_school,
    "Average Reading Score":average_reading_score_school,
    "% Passing Math":percentage_passing_math_school, 
    "% Passing Math":percentage_passing_reading_school,
    "% Overall Passing":overall_passing_rate})


# Formatting
summary_schools["Total School Budget"] =  summary_schools["Total School Budget"].map("${:,.2f}".format)
summary_schools["Per Student Budget"] = summary_schools["Per Student Budget"].map("${:,.2f}".format)

# Display the DataFrame
summary_schools
```

```python
#Highest-Performing Schools (by % Overall Passing)
top_schools=summary_schools.sort_values("% Overall Passing", ascending=False)
top_schools.head(5)
```

```python
#Lowest-Performing Schools (by % Overall Passing)
bottom_schools=summary_schools.sort_values("% Overall Passing")
bottom_schools.head(5)
```

```python
#Math Scores by Grade: Perform the necessary calculations to create a DataFrame that lists the average math score for students of each grade level (9th, 10th, 11th, 12th) at each school.
math_9th=school_data_complete.loc[school_data_complete["grade"]=="9th"].groupby(["school_name"])["math_score"].mean()
math_9th.head()
math_10th=school_data_complete.loc[school_data_complete["grade"]=="10th"].groupby(["school_name"])["math_score"].mean()
math_10th.head()
math_11th=school_data_complete.loc[school_data_complete["grade"]=="11th"].groupby(["school_name"])["math_score"].mean()
math_11th.head()
math_12th=school_data_complete.loc[school_data_complete["grade"]=="12th"].groupby(["school_name"])["math_score"].mean()
math_12th.head()

math_bygrade=pd.DataFrame({
    "9th":math_9th,
    "10th":math_10th,
    "11th":math_11th,
    "12th":math_12th
})
math_bygrade
```

```python
#Reading Scores by Grade: Perform the necessary calculations to create a DataFrame that lists the average reading score for students of each grade level (9th, 10th, 11th, 12th) at each school.
read_9th=school_data_complete.loc[school_data_complete["grade"]=="9th"].groupby(["school_name"])["reading_score"].mean()
read_9th.head()
read_10th=school_data_complete.loc[school_data_complete["grade"]=="10th"].groupby(["school_name"])["reading_score"].mean()
read_10th.head()
read_11th=school_data_complete.loc[school_data_complete["grade"]=="11th"].groupby(["school_name"])["reading_score"].mean()
read_11th.head()
read_12th=school_data_complete.loc[school_data_complete["grade"]=="12th"].groupby(["school_name"])["reading_score"].mean()
read_12th.head()

read_bygrade=pd.DataFrame({
    "9th":read_9th,
    "10th":read_10th,
    "11th":read_11th,
    "12th":read_12th
})
read_bygrade
```

```python
#Scores by School Spending
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
#Create a copy of the school summary since it has the "Per Student Budget"
school_spending_df = summary_schools.copy()
school_spending_df

school_spending_df["Per Student Budget"]=school_spending_df["Per Student Budget"].str.replace("$","").astype("float")
```

```python
# Use `pd.cut` to categorize spending based on the bins.
school_spending_df["Spending Ranges (Per Student)"]=pd.cut(school_spending_df["Per Student Budget"],spending_bins,labels=labels)
school_spending_df
```

```python
#  Calculate averages for every column 
grouped_spending=school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()
grouped_spending
```

```python
#  Keep only desire columns
spending_summary=grouped_spending[["Average Math Score","Average Reading Score","% Passing Math","% Overall Passing"]]
spending_summary
```

```python
#Scores by School Size
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
#Create a copy of the school summary
school_size_df = summary_schools.copy()
school_size_df
```

```python
# Use `pd.cut` to categorize size based on the bins.
school_size_df["School Size"]=pd.cut(school_size_df["Total Students"],size_bins,labels=labels)
school_size_df
```

```python
#  Calculate averages for every column 
grouped_size=school_size_df.groupby(["School Size"]).mean()
grouped_size
```

```python
#  Keep only desire columns
size_summary=grouped_size[["Average Math Score","Average Reading Score","% Passing Math","% Overall Passing"]]
size_summary
```
